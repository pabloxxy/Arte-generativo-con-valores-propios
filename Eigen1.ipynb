```python
import matplotlib.pyplot as plt
import numpy as np
import random

#parametros
n = 10000                                     #número de puntos en la gráfica. 10e5 sale muy rápido pero 10e6 sale con muy buen detalle.
dim = 7                                       #dimensión de la matríz
angle = 0.00                                  #un parámetro para mover las dos entradas variables de la matríz
eigenvalues0 = np.empty(shape=(0))            #un arreglo vacío para iniciar otro
r = random.sample(range(dim), 4)              #4 números aleatorios por si quiero hacer las entradas aleatorias


#Creando la matriz: es una matriz aleatoria con entradas en los complejos. Hay dos factores para alterar la magnitud de la parte real y de la compleja.
random_real_part = np.random.rand(dim, dim)
random_imaginary_part = np.random.rand(dim, dim)
matrix = 0.5* random_real_part + 1j * 0.1 * random_imaginary_part



print(matrix)


#calculando los valores propios e iterando:
for i in range (1, n):

 #escoger si los parámetros son reales o complejos y alterarlos.
 #t = 0.5*(np.cos(angle*1.41+2.7)) + 0.5*(1j * np.sin(angle*1.41+2.7))
 s = 0.5*np.cos(angle) + 1j * 0.5 * np.sin(angle)

 t = 0.5*(np.cos(angle*1.41+2.7))
 #s = 0.5*np.cos(angle)

 #escoger el lugar de los parámetros será aleatorio o será en (7,2) y (6,6)
 matrix [r[0], r[1]] = t
 matrix [r[2], r[3]] = s

 #matrix [6, 1] = t
 #matrix [5, 5] = s

 eigenvalues = np.linalg.eigvals(matrix)                                  #calcular los valores propios de la matriz

 angle += 0.15                                                            #aumentar el parámetro

 eigenvalues0 = np.concatenate((eigenvalues0, eigenvalues), axis=0)       #concatenar los nuevos valores propios al arreglo donde están todos guardados

# Print the eigenvalues
# print("Eigenvalues of the matrix:")
# print(eigenvalues0)

real_part = np.real(eigenvalues0)
imaginary_part = np.imag(eigenvalues0)

# Crear la gráfica en el plano complejo
plt.scatter(real_part, imaginary_part,label = '', color='black', s=0.0005) #s es el tamaño de los puntos y se puede ajustar. Más puntos requerirán menor tamaño.

plt.grid(False)
plt.show()

```
