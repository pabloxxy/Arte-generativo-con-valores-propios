```python
#Un intento de hacer una gráfica en 3d. Tengo 3 listas que tienen las coordenadas x,y,z respectivamente de muchos puntos pero no sé graficarlos en el espacio.
import matplotlib.pyplot as plt
import numpy as np
import random
from matplotlib.animation import FuncAnimation
from IPython.display import HTML
from IPython.display import display
from mpl_toolkits.mplot3d import Axes3D
import plotly.express as px
import pandas as pd

valor_inicial = 0
total_frames = 200
frame = 0

random_real_part = np.random.rand(7, 7)
random_imaginary_part = np.random.rand(7, 7)
matrix = 0.2* random_real_part + 1j * 0.5*random_imaginary_part

print(valor_inicial)
print(matrix)
eigenvalues_total = np.empty(shape=(0))
z_total = np.empty(shape=(0))
real_part_total = np.empty(shape=(0))
imaginary_part_total = np.empty(shape=(0))

for j in range (1, total_frames):

 n = 1000
 eigenvalues0 = np.empty(shape=(0))
 z0 = np.empty(shape=(0))
 angle = 0
 r = np.cos(frame*0.01) + np.sin(frame*0.01)*1j

 for i in range (1, n):

  t = 0.5*(np.cos(angle*1.41+2.7)) + 0.5*(1j * np.sin(angle*1.41+2.7))
  s = 0.5*np.cos(angle) + 1j * 0.5 * np.sin(angle)

  #t = 0.1*(np.cos(angle*1.41+2.7))
  #s = np.sin(angle)

  matrix [6][1] = t
  matrix [5][5] = s
  matrix [1][3] = r

  eigenvalues = np.linalg.eigvals(matrix)

  angle += 0.15
  eigenvalues0 = np.concatenate((eigenvalues0, eigenvalues), axis=0)
  for k in range(1,8):
   z0 = np.concatenate((z0, [frame]), axis=0)

 frame +=1
 real_part = np.real(eigenvalues0)
 imaginary_part = np.imag(eigenvalues0)

 real_part_total = np.concatenate((real_part_total, real_part), axis=0)
 imaginary_part_total = np.concatenate((imaginary_part_total, imaginary_part), axis=0)
 z_total = np.concatenate((z_total, z0), axis=0)
 z_total1 = z_total * (1/7) * (1/n)


df = pd.DataFrame({'real_part_total': real_part_total, 'imaginary_part_total': imaginary_part_total, 'z_total1': z_total1})
fig = px.scatter_3d(df, x='real_part_total', y='imaginary_part_total', z='z_total1', title='Interactive 3D Scatter Plot')

#falta displayear fig

```
